```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('/content/bank_transactions.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.dropna()
```

```python
df = df.sort_values('TransactionDate')
df
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['TransactionDate']
```

```python
df['Month'] = df['TransactionDate'].dt.month
df['Month']
```

```python
monthly_spending = df.groupby('Month')['TransactionAmount (INR)'].sum()
print(monthly_spending)
```

```python
df.groupby('TransactionType')['Amount'].sum()
```

```python
total_transactions = df['TransactionID'].nunique()
print("Total transactions:", total_transactions)
```

```python
transactions_per_month = df.groupby('Month')['TransactionID'].count()
print(transactions_per_month)
```

```python
# Make sure TransactionDate is datetime
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])

# Extract month
df['Month'] = df['TransactionDate'].dt.month

# Total inflow (Amount > 0)
monthly_inflow = df[df['Amount'] > 0].groupby('Month')['Amount'].sum()

# Total outflow (Amount < 0)
monthly_outflow = df[df['Amount'] < 0].groupby('Month')['Amount'].sum()

print("Monthly Inflow:\n", monthly_inflow)
print("Monthly Outflow:\n", monthly_outflow)
```

```python
monthly_inflow = df[df['TransactionAmount (INR)'] > 0].groupby('Month')['TransactionAmount (INR)'].sum()
monthly_outflow = df[df['TransactionAmount (INR)'] < 0].groupby('Month')['TransactionAmount (INR)'].sum
print("Monthly Inflow:\n", monthly_inflow)
print("Monthly Outflow:\n", monthly_outflow)
```

```python
df.nlargest(5, 'TransactionAmount (INR)')
```

```python
# Find row with max amount
max_transaction = df.loc[df['TransactionAmount (INR)'].idxmax()]

print(max_transaction)
```

```python
person_highest = df.loc[df['TransactionAmount (INR)'].idxmax(), 'CustomerID']
print("Person with highest transaction:", person_highest)
```

```python
plt.figure(figsize=(8,5))
monthly_spending.plot(kind='line', marker='o',color='purple')
plt.title("Monthly Transaction Trend")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.grid(True)
plt.show()
```

```python
category_spending = df.groupby('CustomerID')['TransactionAmount (INR)'].sum()
category_spending.plot(kind='bar', color='skyblue')
plt.title("Spending by CustomerID")
plt.xlabel("CustomerID")
plt.ylabel("Total Amount")
plt.show()
```

```python
top5 = df.nlargest(5, 'TransactionAmount (INR)')

plt.figure(figsize=(8,5))
sns.barplot(x='TransactionAmount (INR)', y='TransactionID', data=top5, palette='magma')
plt.title("Top 5 Highest Transactions")
plt.xlabel("TransactionAmount (INR)")
plt.ylabel("TransactionID")
plt.show()
```

```python
top5 = df.nlargest(5, 'TransactionAmount (INR)')

plt.figure(figsize=(8,5))
sns.barplot(x='TransactionAmount (INR)', y='TransactionID', data=top5, palette='coolwarm')
plt.title("Top 5 Highest Transactions")
plt.xlabel("Amount")
plt.ylabel("Transaction ID")
plt.show()
```

```python
inflow = df[df['TransactionAmount (INR)'] > 0]['TransactionAmount (INR)'].sum()
outflow = abs(df[df['TransactionAmount (INR)'] < 0]['TransactionAmount (INR)'].sum())
plt.figure(figsize=(6,6))
plt.pie([inflow, outflow], labels=['Inflow', 'Outflow'], autopct='%1.1f%%', colors=['skyblue','red'])
plt.title("Inflow vs Outflow")
plt.show()
```
